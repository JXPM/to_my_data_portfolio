Je vais identifier les outliers et les traités 

```python
Q1_link = Amazon['product_link'].quantile(0.25)
Q3_link = Amazon['product_link'].quantile(0.75)
IQR_link = Q3_link - Q1_link
lower_link = Q1_link - 1.5 * IQR_link
upper_link = Q3_link + 1.5 * IQR_link

Q1_name = Amazon['rating'].quantile(0.25)
Q3_name = Amazon['rating'].quantile(0.75)
IQR_name = Q3_name - Q1_name
lower_name = Q1_name - 1.5 * IQR_name
upper_name = Q3_name + 1.5 * IQR_name

Amazon['is_outlier'] = (
    (Amazon['product_link'] < lower_link) | (Amazon['product_link'] > upper_link) |
    (Amazon['rating'] < lower_name) | (Amazon['rating'] > upper_name)
)

print("BORNES POUR product_link :")
print(f"Lower bound: {lower_link}, Upper bound: {upper_link}")
print("BORNES POUR product_name :")
print(f"Lower bound: {lower_name}, Upper bound: {upper_name}")
print(Amazon[['product_link', 'rating', 'is_outlier']])
```

Je ne sais pourquoi, mais le calcule d'outliers ne detecte aucun outliers 

Je vais donc essayer la methode Z-Score afin d'examiner le problème 

```python
from scipy.stats import zscore

Amazon['zscore_product_link'] = zscore(Amazon['product_link'])
Amazon['zscore_rating'] = zscore(Amazon['rating'])

Amazon['is_outlier'] = (
    (np.abs(Amazon['zscore_product_link']) > 3) | 
    (np.abs(Amazon['zscore_rating']) > 3)
)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='product_link',
    y='rating',
    data=Amazon,
    hue='is_outlier',
    palette={False: 'blue', True: 'red'}
)

plt.title('Scatter plot of product link and rating\n(Outliers in Red)')
plt.xlabel('product_link')
plt.ylabel('rating')
plt.legend(title='Outlier', labels=['Non-Outlier', 'Outlier'])
plt.show()
```
